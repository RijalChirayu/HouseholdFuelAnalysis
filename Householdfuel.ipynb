### Renaming the columns

```python
df.rename(columns={'Table 07: Number of households by type of fuel usually used for cooking, NPHC 2021':'Households by fuel type',
       'Unnamed: 1':'Region', 'Unnamed: 2':'Total', 'Unnamed: 3':'Firewood', 'Unnamed: 4':'LPG', 'Unnamed: 5': 'Electricity',
       'Unnamed: 6' : 'Cow dung', 'Unnamed: 7': 'Bio Gas', 'Unnamed: 8':'Kerosense', 'Unnamed: 9':'Others'} , inplace = True)
```

Selecting data without the first column

```python
data = df.drop(['Households by fuel type'] , axis = 1)
```

Selecting Certain Urban and Rural Region

```python
region_data = data.loc[7:9]
region_data
```

Plotting the data for households using Firewood , LPG and Electricity

```python
region_data.plot(kind='bar' , x = 'Region' , y = ['Firewood' , 'LPG' , 'Electricity'] , stacked= False , rot= 0 , ylabel='No of fuel used (in 10 lakhs) ' , title = 'Fuel used in Rural and Urban region of Nepal')
```
